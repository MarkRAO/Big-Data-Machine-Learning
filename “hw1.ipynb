```
#Q6
import jax.numpy as np
from jax import grad
from jax.nn import relu,sigmoid
import matplotlib.pyplot as plt
x = np.linspace(-10,10,1000)
dr = np.vectorize(grad(relu))
ds = np.vectorize(grad(sigmoid))
dt = np.vectorize(grad(np.tanh))
fig,axs = plt.subplots(1,3,figsize=(12,4),constrained_layout=True)
fig.suptitle('gradient plot')
axs[0].plot(x,sigmoid(x),label='sigmoid')
axs[0].plot(x,ds(x),label='sigmoid-grad')
axs[0].legend()
axs[1].plot(x,relu(x),label='relu')
axs[1].plot(x,dr(x),label='relu-grad')
axs[1].legend()
axs[2].plot(x,np.tanh(x),label='tanh')
axs[2].plot(x,dt(x),label='tanh-grad')
axs[2].legend()
```
